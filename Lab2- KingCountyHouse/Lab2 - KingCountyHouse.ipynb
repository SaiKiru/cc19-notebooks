## Cleaning Data

### Removing Rows with NaN values

```python
# I'm only removing rows that I don't know
# what to do about the NaN values
df = df.dropna(subset=[
    'id',
    'sqft_living',
    'sqft_lot',
    'floors',
    'waterfront',
    'view',
    'sqft_above',
    'sqft_basement',
    'zipcode',
    'sqft_lot15'
])
df.isnull().sum()
```

### Removing Duplicates

```python
# assuming that the ID is supposed to be unique...
df = df.drop_duplicates(subset='id')
df.duplicated(subset='id').sum()
```

### Cleaning `date`

```python
# replace NaN dates with November 1, 2014,
# which is the midpoint of the testing period
df['date'] = df['date'].apply(
    lambda x:
        datetime.datetime(2014, 11, 1).strftime('%Y%m%dT%H%M%S')
        if (pd.isnull(x))
        
        else
        x
)
df['date'].isnull().sum()
```

### Cleaning `price`

```python
# Convert negative prices to positive
df['price'] = df['price'].apply(
    lambda x:
        abs(x)
        if not (pd.isnull(x))

        else
        x
)

df['price'].describe()
```

```python
# if price is NaN or 0, change it to the mean price
df['price'] = df['price'].apply(
    lambda x:
        df['price'].mean()
        if (x == 0)

        else
        x
)

df['price'] = df['price'].apply(
    lambda x:
        df['price'].mean()
        if (pd.isnull(x))

        else
        x
)

df['price'].describe()
```

### Cleaning `bedroom`
because my mother said so

```python
# if bedrooms is NaN or <1, convert it to mean
df['bedrooms'] = df['bedrooms'].apply(
    lambda x:
        df['bedrooms'].mean()
        if (x < 1)

        else
        x
)

df['bedrooms'] = df['bedrooms'].apply(
    lambda x:
        df['bedrooms'].mean()
        if pd.isnull(x)

        else
        x
)

df['bedrooms'].describe()
```

```python
# convert data type to int
# because there's no such thing as 2.2 bedrooms
df['bedrooms'] = df['bedrooms'].astype(int)
df['bedrooms'].dtypes
```

### Cleaning `bathroom`
because my father said so

```python
# if bathrooms is NaN or <1, convert it to mean
df['bathrooms'] = df['bathrooms'].apply(
    lambda x:
        df['bathrooms'].mean()
        if (x < 1)

        else
        x
)

df['bathrooms'] = df['bathrooms'].apply(
    lambda x:
        df['bathrooms'].mean()
        if pd.isnull(x)

        else
        x
)

df['bathrooms'].describe()
```

```python
# convert date type to int
# because there's no such thing as 2.2 bathrooms
df['bathrooms'] = df['bathrooms'].astype(int)
df['bathrooms'].dtypes
```

### Cleaning `condition`

```python
# if condition is NaN, convert it to mean
df['condition'] = df['condition'].apply(
    lambda x:
        df['condition'].mean()
        if pd.isnull(x)

        else
        x
)

df['condition'].isnull().sum()
```

### Cleaning `lat` and `long`

![](https://i.imgur.com/Ajdcwgc.png) \
_A quick search of the coordinates of King County._

Latitudes must be positive \
Longitudes must be negative

```python
# convert latitudes to positive
# and longitudes to negative
df['lat'] = df['lat'].apply(lambda x: abs(x))
df['long'] = df['long'].apply(lambda x: abs(x) * -1)

df[['lat', 'long']].describe()
```

### Cleaning `sqft_living`

```python
# converting data type to float
# because it's an area measurement
# well it doesn't really matter because the precision is already lost
# but it bothers me
df['sqft_living'] = df['sqft_living'].astype(float)
df['sqft_living'].dtypes
```
