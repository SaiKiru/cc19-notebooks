## Cleaning Data

### Removing Duplicates

```python
# assuming that the ID is supposed to be unique,
# I'm checking the values of rows with duplicate ID
# because I noticed something odd and my instincts are calling out to me

mask = df['id'].duplicated(keep=False)
duplicates = df[mask]

submask = duplicates.iloc[:, 4:].duplicated(keep=False)
inconsistent_duplicates = duplicates[~submask]
```

### Cleaning `date`

```python
# replace NaN dates with November 1, 2014,
# which is the midpoint of the testing period
df[df['date'].isnull()] = datetime.datetime(2014, 11, 1).strftime('%Y%m%dT%H%M%S')
```

```python
# convert latitudes to positive
# and longitudes to negative
df['lat'] = df['lat'].apply(lambda x: abs(x))
df['long'] = df['long'].apply(lambda x: abs(x) * -1)

df[['lat', 'long']].describe()
```
